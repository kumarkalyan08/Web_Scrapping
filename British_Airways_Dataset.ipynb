```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
from urllib.parse import urljoin
```

```python
headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}
```

```python
review_list=[]
url= 'https://www.airlinequality.com/airline-reviews/british-airways/page/1/'
i=1
max_iter=2
while i<=max_iter:
    print(url)
    response= requests.get(url,headers=headers)
    webpage_content=BeautifulSoup(response.content,'html.parser')
    all_reviews=webpage_content.find_all('div',class_='body')
    
    
    for each_review in all_reviews:
        reviews_dict={}
        reviews_dict['exp_in_breif']=each_review.h2.text.strip()
        reviews_dict['Name']=each_review.span.text.strip()
        if len(each_review.find('div',class_='text_content').text.strip().split('|'))>1:
            reviews_dict['Trip_Verified']=each_review.find('div',class_='text_content').text.strip().split('|')[0]
            reviews_dict['content']= each_review.find('div',class_='text_content').text.strip().split('|')[1]
        else:
            reviews_dict['content']= each_review.find('div',class_='text_content').text.strip().split('|')[0]
            reviews_dict['Trip_Verified']= 'Nan'
                   
        more_details=[other for other in each_review.find_all('tr')]
        for record in more_details:
            first_td= record.find('td')
            if first_td.find_next_sibling('td',class_='review-rating-stars stars'):
                founded_record=first_td.find_next_sibling('td',class_='review-rating-stars stars')
                star_fill=founded_record.find_all('span',class_='star fill')
                for max in star_fill:
                    second=max
            else:
                second=first_td.find_next_sibling('td')  
                
            if(first_td and second):
                #print(first_td.text.strip(),second.text.strip())
                reviews_dict[str(first_td.text.strip())]=second.text.strip()
            
            else:
                continue
        review_list.append(reviews_dict)
    #i=i+1
    total_page=webpage_content.find_all('section',class_='layout-section layout-2 closer-top')
    for x in total_page:
        i=int((x.select('a.active')[0].text.strip()))+1
    #print(review_list)
    print("length",len(review_list))
    url=urljoin('https://www.airlinequality.com/airline-reviews/british-airways/page/1',str(i))
    
    length_list=[]
    for x in total_page:
        all_a_tags=x.select('li a')
    for each_a in all_a_tags:
        length_list.append(each_a.text.strip())
    numerical_numbers=[int(x) for x in length_list if x.isdigit()]
    k=1
    for num in numerical_numbers:
        if num>k:
            max_iter=num
        else:
            max_iter=k
    #print(numerical_numbers)
    #print(max_iter)
    #url= urljoin(url,i)
```
